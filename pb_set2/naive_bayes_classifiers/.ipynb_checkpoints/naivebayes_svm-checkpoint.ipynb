```python
def vectorize(x):
    vocabulary={}
    count=0
    for i in set(x):
        if ~(i in vocabulary.values()):
            vocabulary[count]=i
            count+=1
    return vocabulary
```

```python
def transform(x):
    array=[0 for _ in range(x.shape[1])]
    for i,_ in enumerate(x):
        if x[i] in vocabulary.values():
            array[i]=1
    return array
```
