```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train=np.genfromtxt('train.csv',delimiter=',',skip_header=True)
valid=np.genfromtxt('valid.csv',delimiter=',',skip_header=True)
```

```python
x_train=train[:,:-1]
y_train=train[:,-1:]
intercept=np.ones((x_train.shape[0],1))
x_train=np.append(intercept,x_train,axis=1)
```

```python
def perceptron(x,y):
    theta=np.zeros((x.shape[1],1))
    lr=0.01
    d=0
    for i in range(x.shape[0]):
        d=1 if y[i][0]==1 else 0
        train_exm=x[i,:].reshape(x.shape[1],1)
        theta=theta+lr*d*train_exm
    return theta
```

```python
theta=perceptron(x_train,y_train)
```

```python
theta
```

```python
def split_plot_p(x,y):
    positive_x=x[y[:,0]>=0.5]
    negative_x=x[y[:,0]<=0.5]
    x1_p=positive_x[:,1]
    x2_p=positive_x[:,2]
    x1_n=negative_x[:,1]
    x2_n=negative_x[:,2]
    return x1_p,x2_p,x1_n,x2_n
```

```python
x1_p,x2_p,x1_n,x2_n=split_plot_p(x_train,y_train)
```

```python
x1_line=[i for i in range(-6,6)]
a=theta[1][0]
b=theta[2][0]
```

```python
plt.figure(figsize=(14,6))
plt.scatter(x1_p,x2_p,color='green')
plt.scatter(x1_n,x2_n,color='red')
plt.plot(x1_line, [(-a/b)*x for x in x1_line],color='blue')
plt.show()
```

### Perceptron on validate set

```python
x_valid=valid[:,:-1]
y_valid=valid[:,-1:]
intercept=np.ones((x_valid.shape[0],1))
x_valid=np.append(intercept,x_valid,axis=1)
```

```python
x1_p,x2_p,x1_n,x2_n=split_plot_p(x_valid,y_valid)
```

```python
plt.figure(figsize=(14,6))
plt.scatter(x1_p,x2_p,color='green')
plt.scatter(x1_n,x2_n,color='red')
plt.plot(x1_line, [(-a/b)*x for x in x1_line],color='blue')
plt.show()
```
